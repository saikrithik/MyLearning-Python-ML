# IGN game review data analysis

Read the ign.csv file and find the following:
1. Number of games having more than 7 score
2. How many got the editor choice along with year and more conditions
3. Which genre has the maximum score
4. Which game has the minimum score
5. Which platform has the highest score
6. Average score of xbox platform
7. Which year had the most releases
8. How many games were released for both Xbox 360 and PC
Also plot the following using pandas plot function:
1. Bar graph of score category
2. Bar graph of editor choice
3. Bar graph of genre


```python
import pandas as pd
df = pd.read_csv("./ign.csv")
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df
```

1. Number of games having more than 7 score


```python
gg=df[df['score']>7.0]
print('there are',gg['score'].count(),'games with more than 7 score')
gg
```

2. How many got the editor choice along with year and more conditions


```python
ec=df[df['editors_choice']=='Y']
print('there are',ec['editors_choice'].count(),' games as editors choice')
ec.loc[:3517,'editors_choice':'release_day']
```

3. Which genre has the maximum score


```python
gs=df[(df['score']==10.0)]
gs1=gs['genre'].value_counts()
gs1.index
```

4. Which game has the minimum score


```python
gs=df[(df['score']==0.5)]
gs1=gs['title'].value_counts()
gs1.head(1)
```

5. Which platform has the highest score


```python
gs=df[(df['score']==10)]
gs1=gs['platform'].value_counts()
gs1.head(1)
```

6. Average score of xbox platform


```python
avg=df[(df['platform']=='Xbox') | (df['platform']=='Xbox 360')]
avg['score'].mean()
```

7. Which year had the most releases


```python
int(df['release_year'].mean())
```

8. How many games were released for both Xbox 360 and PC

```python
df[(df['platform']=='Xbox 360') | (df['platform']=='PC')].groupby('title').count().index
# for x,y in ec.groupby('title'):
#     print(x)
#     print(y)
```

1. Bar graph of score category


```python
df["score"]=df.score.astype(int)
df['score'].value_counts().plot(kind='bar')
```

2. Bar graph of editor choice


```python
import matplotlib.pyplot as plt
df['editors_choice'].value_counts().plot(kind='bar')
```
