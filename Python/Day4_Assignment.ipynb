# LA-8 Inverse of matrix

```python
#### import numpy as np
k=3
a = np.zeros((k,k),dtype = int)
print("enter {0} elements".format(k*k))
for i in range(k):
    for j in range(k):
        a[i][j]=input()
print(a)
print("inverse of the matrix=\n",a.T/np.linalg.det(a))
```

# LA-7 Number of occurences

```python
f=open('LotsOFWords.txt', 'r')
n=0
for l in f:
    w = l.split()
    for i in range(len(w)):
        if (w[i]=='apple' or w[i]=='apple.' or w[i]=='apple?'):
            n=n+1
print(n)
```

# LA-9 Distance between two points

```python
import numpy as np
a = np.zeros((1,2),dtype = int)
print("enter 2 elements")
for i in range(1):
    for j in range(2):
        a[i][j]=int(input())
#a=a.ravel()
b = np.zeros((1,2),dtype = int)
print("enter 2 elements")
for i in range(1):
    for j in range(2):
        b[i][j]=int(input())
#b=b.ravel()
sq1=pow((a[0:2,0]-b[0:2,0]),2)
sq2=pow((a[0:2,1]-b[0:2,1]),2)
sq=pow((sq1+sq2),1/2)
print(sq)
```
